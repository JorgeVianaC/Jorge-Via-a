```python
class Personaje:
    def __init__(self, rol: str, elemento: str, nivel: int, vida: int, pa: int, pm: int) -> None:
        self.rol: str = rol
        self.elemento: str = elemento
        self.nivel: int = nivel
        self.vida: int = vida
        self.pa: int = pa
        self.pm: int = pm

    def inform(self) -> str:
        return f"{self.rol}\nElemento {self.elemento}, nivel {self.nivel}, P.Vida {self.vida}, P.Accion {self.pa}, P.Movimiento {self.pm}"

    def mover(self, moverse: int) -> None:
        self.pm -= moverse

    def actuar(self, accion: int) -> None:
        self.pa -= accion

class Ocra(Personaje):
    def inform(self) -> str:
        return "Ocra: " + super().inform() +"\n"

    def ataqueocra(self) -> None:
        self.actuar(4)

class Yopuka(Personaje):
    def inform(self) -> str:
        return "Yopuka: " + super().inform() +"\n"

    def ataqueyopuka(self) -> None:
        self.actuar(5)

class Aniripsa(Personaje):
    def inform(self) -> str:
        return "Aniripsa: " + super().inform() +"\n"

    def ataqueaniripsa(self) -> None:
        self.actuar(4)
```

```python
pj1 = Ocra(rol="Ataque a distancia", elemento="Fuego", nivel=200, vida=4300, pa=11, pm=6)
pj2 = Yopuka(rol = "Combate cercano", elemento="Tierra", nivel=200, vida=4800, pa=12, pm=5)
pj3 = Aniripsa(rol = "Support", elemento="Fuego", nivel=200, vida=4500, pa=11, pm=6)
personaje: list[Personaje] = [pj1, pj2, pj3]

pj1.mover(3)
pj1.ataqueocra()

pj2.mover(5)
pj2.ataqueyopuka()

pj3.mover(4)
pj3.ataqueaniripsa()
pj3.ataqueaniripsa()

for p in personaje:
    print(p.inform())


```
