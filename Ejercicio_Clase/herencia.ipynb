```python
class Personaje:
    def __init__(self, elemento: str, nivel: int, vida: int, pa: int, pm: int) -> None:
        self.elemento: str = elemento
        self.nivel: int = nivel
        self.vida: int = vida
        self.pa: int = pa
        self.pm: int = pm

    def inform(self) -> str:
        return f"Elemento {self.elemento}, nivel {self.nivel}, P.Vida {self.vida}, P.Accion {self.pa}, P.Movimiento {self.pm}"

class Ocra(Personaje):
    def inform(self) -> str:
        rol = "Combate a distancia"
        return "Ocra: " + rol +"\n" + super().inform() +"\n"

class Yopuka(Personaje):
    def inform(self) -> str:
        rol = "Combate cercano"
        return "Yopuka: " + rol +"\n" + super().inform() +"\n"

class Aniripsa(Personaje):
    def inform(self) -> str:
        rol = "Support"
        return "Aniripsa: " + rol +"\n" + super().inform() +"\n"
```

```python
pj1 = Ocra(elemento="Fuego", nivel=200, vida=4300, pa=11, pm=6)
pj2 = Yopuka(elemento="Tierra", nivel=200, vida=4800, pa=12, pm=5)
pj3 = Aniripsa(elemento="Fuego", nivel=200, vida=4500, pa=11, pm=6)
personaje: list[Personaje] = [pj1, pj2, pj3]

for p in personaje:
    print(p.inform())
```
